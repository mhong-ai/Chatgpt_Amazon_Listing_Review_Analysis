````python
#Import dataset from google drive;
# To check 10 rows of the datasets
import pandas as pd
df = pd.read_csv('/content/gdrive/My Drive/my_projects/chatgpt/Birkenstock-1.csv')
df.head(10)
````

````python
#To separate "Style" column into "Size", and "Color" column based on "|" separator
df[['Size', 'Color1']] = df['Style'].str.split('|', expand=True)
````

````python
df = df.drop('Style', axis=1)
````

````python
#To separate "Color" column into "Color1", and "Color" column based on ":" separator, then drop "Color1" coloum
df[['Color1', 'Color']] = df['Color1'].str.split(':', expand=True)
df = df.drop('Color1', axis=1)
````

````python
df.head(10)
````

````python
#To separate "Date" column into "Reviewed Location", and "Date2" column based on ":" separator, then drop "Date" coloum
df[['Reviewed Location', 'Date2']] = df['Date'].str.split('on', expand=True)
df = df.drop('Date', axis=1)
````

````python
# Rename the 'Date2' column to 'Date'
df = df.rename(columns={'Date2': 'Date'})
````

````python
df.head(10)
````

````python
# Create a new column 'Review' with sequential values

df['Review'] = [f'review_{i}' for i in range(1, len(df) + 1)]
````

````python
# Move the "Review" column to column 5
review_column = df.pop('Review')  # Remove the "Review" column from its current position
df.insert(4, 'Review', review_column)  # Insert the "Review" column at column 5
````

````python
# Combine the "Review" and "Body" columns into a new column "Reviews"
df['Reviews'] = df['Review'] + '=' + df['Body']
````
